```python
eventsS = {}
eventsS[0] = '0Bs'
eventsS[1] = '1Bs'
eventsS[2] = '2Bs'
eventsS[3] = '1B2Bs'
eventsS[4] = '3Bs'
eventsS[5] = '1B3Bs'
eventsS[6] = '2B3Bs'
eventsS[7] = 'LOADs'

eventsE = {}
eventsE[0] = '0Bs'
eventsE[1] = '1Bs'
eventsE[2] = '2Bs'
eventsE[3] = '1B2Bs'
eventsE[4] = '3Bs'
eventsE[5] = '1B3Bs'
eventsE[6] = '2B3Bs'
eventsE[7] = 'LOADs'

print(eventsS)
print(eventsE)
```

```python
dfM = df#.head(1000)
totals = {}
totals['outs'] = sum(dfM['outsOnPlay'])
totals['runs'] = sum(dfM['runsOnPlay'])
totals['0Bs'] = sum(dfM['START_BASES_CD']==0)
totals['1Bs'] = sum(dfM['START_BASES_CD']==1)
totals['2Bs'] = sum(dfM['START_BASES_CD']==2)
totals['3Bs'] = sum(dfM['START_BASES_CD']==4)
totals['1B2Bs'] = sum(dfM['START_BASES_CD']==3)
totals['1B3Bs'] = sum(dfM['START_BASES_CD']==5)
totals['2B3Bs'] = sum(dfM['START_BASES_CD']==6)
totals['LOADs'] = sum(dfM['START_BASES_CD']==7)
totals['0Be'] = sum(dfM['END_BASES_CD']==0)
totals['1Be'] = sum(dfM['END_BASES_CD']==1)
totals['2Be'] = sum(dfM['END_BASES_CD']==2)
totals['3Be'] = sum(dfM['END_BASES_CD']==4)
totals['1B2Be'] = sum(dfM['END_BASES_CD']==3)
totals['1B3Be'] = sum(dfM['END_BASES_CD']==5)
totals['2B3Be'] = sum(dfM['END_BASES_CD']==6)
totals['LOADe'] = sum(dfM['END_BASES_CD']==7)


print(totals)
```

```python
transitions = pd.DataFrame(index=range(8),columns=range(8))
# transitions['totals'] = 0
for i in range(len(transitions)):
    transitions[i] = 0
    
# should be matrix of zeros
# print(transitions)

# clever way of adding transitions
for i in range(len(dfM['GAME_ID'])):
    # might need to add an "outs condition; especially down the diagonal"
    t1 = dfM.at[i,'START_BASES_CD']
    t2 = dfM.at[i,'END_BASES_CD']
    transitions.loc[t1,t2] += 1

    
# check for raw numbers:
print(transitions)


#finalize PROBABILITY TRANSITION MATRIX:
for i in range(len(transitions)):
    transitions.iloc[i] = transitions.iloc[i] / totals[eventsS[i]]
#     print(transitions.iloc[i])
#     print("sumrow" +str(sum(transitions.iloc[i])))

        
print(transitions)
```

```python
# -- CHECKING IF ALGO RUNS CORRECTLY
threshold = 50
blU = pd.DataFrame(dfM.player.unique()) #I don't think I need this threshold
blU.columns = ['player']
blU['outsFactor'] = 0
blU['runsFactor'] = 0
blU['baseFactor'] = 0
blU['tot'] = 0

for batter in blU['player']:
    temp = dfM[dfM['player']==batter]
    temp = temp.reset_index(drop=True)
    if(len(temp) > threshold):
        for i in range(len(temp['GAME_ID'])):
            # -- OUTS
            if(temp['outsOnPlay'][i] > 0):
                blU.loc[blU['player'] == batter,'outsFactor'] += (temp['outsOnPlay'][i]*(1/totals['outs']))
            # -- RUNS
            if(temp['runsOnPlay'][i] > 0):
                blU.loc[blU['player'] == batter,'runsFactor'] += (temp['runsOnPlay'][i]*(1/totals['runs']))
            # -- BASES   
            if(temp['START_BASES_CD'][i] == temp['END_BASES_CD'][i]):
                blU.loc[blU['player'] == batter,'baseFactor'] += (temp['START_BASES_CD']/totals[eventsS[temp['START_BASES_CD'][i]]])

            if(temp['START_BASES_CD'][i] != temp['END_BASES_CD'][i]):
                blU.loc[blU['player'] == batter,'baseFactor'] += ((np.log(temp['START_BASES_CD']+1)/totals[eventsS[temp['START_BASES_CD'][i]]]) + (np.log(temp['END_BASES_CD']+1)/totals[eventsS[temp['END_BASES_CD'][i]]]))

    else:
        blU = blU[blU.player != batter] #blU.drop[blU['player']== batter]
    
```

```python
#cleanup for beauty, and hopefully a database ;)
blU['baseFactor'] = blU['baseFactor'].fillna(0)            
blU['tot'] = blU['runsFactor']-blU['outsFactor']+blU['baseFactor']
blU = blU.sort_values(by=['tot'],ascending=False)
blU = blU.reset_index(drop=True)
print(blU)
```

```python
blU.rename(columns={'tot': 'mSHS'}, inplace=True)
blU['player'].head(5)
```
